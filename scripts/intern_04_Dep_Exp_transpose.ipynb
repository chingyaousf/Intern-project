```python
#Transpose depency and save to dependency_transposed.csv

# Read the CSV file
dependency = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/22Q2 CRISPR_gene_dependency.csv")

# Transpose the DataFrame
dependency_transposed = dependency.transpose()

# Reset the index and set new column names
dependency_transposed = dependency_transposed.reset_index()
dependency_transposed.columns = dependency_transposed.iloc[0]
dependency_transposed = dependency_transposed.iloc[1:]

# Adjust the leftmost column to count from 0
dependency_transposed.reset_index(drop=True, inplace=True)

# Save the transposed DataFrame to a CSV file
dependency_transposed.to_csv("/content/drive/MyDrive/Colab Notebooks/dependency_transposed.csv", index=False)

dependency_transposed
```

```python
#Transpose gene_expression and save to gene_expression_transposed.csv

# Read the CSV file
gene_expression = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/22Q2 CCLE_expression.csv")

# Transpose the DataFrame
gene_expression_transposed = gene_expression.transpose()

# Reset the index and set new column names
gene_expression_transposed = gene_expression_transposed.reset_index()
gene_expression_transposed.columns = gene_expression_transposed.iloc[0]
gene_expression_transposed = gene_expression_transposed.iloc[1:]

# Adjust the leftmost column to count from 0
gene_expression_transposed.reset_index(drop=True, inplace=True)

# Change the value in the "Unnamed: 0" column to "DepMap_ID"
gene_expression_transposed.rename(columns={"Unnamed: 0": "DepMap_ID"}, inplace=True)

# Save the transposed DataFrame to a CSV file
gene_expression_transposed.to_csv("/content/drive/MyDrive/Colab Notebooks/gene_expression_transposed.csv", index=False)

gene_expression_transposed
```
