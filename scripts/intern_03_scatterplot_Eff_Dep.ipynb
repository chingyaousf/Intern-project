```python
import pandas as pd
```

```python
# Read in the 22Q4 CRISPR_gene_effect.csv
effect = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/22Q2 CRISPR_gene_effect.csv")
effect
```

```python
# Read in the 22Q4 CRISPRGeneDependency.csv
dependency = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/22Q4 CRISPRGeneDependency.csv")
dependency
```

```python
# Merge the two files based on the cell line ID (after merged only keep the common row names)
merged = pd.merge(effect, dependency, left_on='DepMap_ID', right_on='ModelID')
merged
```

```python
# Generate scatter plots for the first 10 matching column pairs

import matplotlib.pyplot as plt

# Get the list of columns to plot (exclude "DepMap_ID" and the "ModelID")
effect_cols = [col[:-2] for col in merged.columns[1:11]]
dependency_cols = [col[:-2] for col in merged.columns[17387:17400]]

# Generate scatter plots for all columns with matching names
for col in effect_cols:
    if col in dependency_cols:
        plt.scatter(merged[col+"_x"], merged[col+"_y"])
        plt.xlabel(col+" effect")
        plt.ylabel(col+" dependency")
        plt.show()
```
