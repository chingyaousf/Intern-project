```python
# To extract the diagonal values bigger than 0.1 and create a new dataset preserving the rows and columns with diagonal values bigger than 0.1

import pandas as pd
import numpy as np

# Read the CSV file into a DataFrame
correlation_table = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/correlation_table_test.csv', index_col=0)

# Get the diagonal values
diagonal_values = np.diag(correlation_table.values)

# Create a boolean mask for diagonal values bigger than 0.1
mask = diagonal_values > 0.1

# Select rows and columns based on the mask
correlation_table_test_diavalues_gt_0 = correlation_table.iloc[mask, mask]

# Save the new dataset to a CSV file in the same directory
correlation_table_test_diavalues_gt_0.to_csv('/content/drive/MyDrive/Colab Notebooks/correlation_table_test_diavalues_gt_0.csv')

# Print the new dataset
correlation_table_test_diavalues_gt_0
```
