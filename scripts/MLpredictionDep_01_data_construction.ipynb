```python
import pandas as pd

# Read the CSV file
arranged_table = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/arranged_table.csv")

# Transpose the DataFrame
arranged_table_transposed = arranged_table.transpose()

# Reset the index and set new column names
arranged_table_transposed = arranged_table_transposed.reset_index()
arranged_table_transposed.columns = arranged_table_transposed.iloc[0]
arranged_table_transposed = arranged_table_transposed.iloc[1:]

# Adjust the leftmost column to count from 0
arranged_table_transposed.reset_index(drop=True, inplace=True)

# Save the transposed DataFrame to a CSV file
arranged_table_transposed.to_csv("/content/drive/MyDrive/Colab Notebooks/DeepDEP/arranged_table_transposed.csv", index=False)

arranged_table_transposed
```

```python
import pandas as pd

# Read the CSV file
arranged_table = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DeepDEP/arranged_table_transposed.csv")

# Separate the column names with "Exp" and "Dep"
exp_columns = [col for col in arranged_table.columns if col.endswith(" Exp")]
dep_columns = [col for col in arranged_table.columns if col.endswith(" Dep")]

# Reorder the columns with "Exp" on the left and "Dep" on the right
new_columns = ["DepMap_ID"] + exp_columns + dep_columns

# Create a new DataFrame with the reordered columns
arranged_table_reordered = arranged_table[new_columns]

# Adjust the leftmost column to count from 0
arranged_table_reordered.reset_index(drop=True, inplace=True)

arranged_table_reordered
```

```python
# Find the column names containing the specified strings
target_columns = arranged_table_reordered.columns[arranged_table_reordered.columns.str.contains('EIF2AK3|SF3B1|NQO1|SQSTM1|MYC|HSP70')]

target_columns
```

```python
# Create a list of column names to preserve

preserved_columns = ["EIF2AK3", "SF3B1", "NQO1", "SQSTM1", "MYC", "HSP70"]

# Filter the columns based on the preserved column names
filtered_columns = [col for col in arranged_table_reordered.columns if (col.endswith(" Dep") and col.split(" ")[0] in preserved_columns) or not col.endswith(" Dep")]

# Create a new DataFrame with the filtered columns
filtered_arranged_table = arranged_table_reordered[filtered_columns]

# Save the filtered_arranged_table DataFrame to a CSV file
filtered_arranged_table.to_csv("/content/drive/MyDrive/Colab Notebooks/DeepDEP/filtered_arranged_table.csv", index=False)

filtered_arranged_table
```

```python
## From 'EIF2AK3 (9451) Dep', 'MYC (4609) Dep', 'NQO1 (1728) Dep', 'SF3B1 (23451) Dep', 'SQSTM1 (8878) Dep_gt_01.csv which extract the all values bigger than 0.1 and then file filtered_arranged_table.csv mach gene EXP with it to produce a dataset for ML.

import pandas as pd

# Define the input file paths for the first files
first_file_paths = [
    '/content/drive/MyDrive/Colab Notebooks/correlation_table_filter_all/EIF2AK3_(9451)_Dep_gt_01.csv',
    '/content/drive/MyDrive/Colab Notebooks/correlation_table_filter_all/MYC_(4609)_Dep_gt_01.csv',
    '/content/drive/MyDrive/Colab Notebooks/correlation_table_filter_all/NQO1_(1728)_Dep_gt_01.csv',
    '/content/drive/MyDrive/Colab Notebooks/correlation_table_filter_all/SF3B1_(23451)_Dep_gt_01.csv',
    '/content/drive/MyDrive/Colab Notebooks/correlation_table_filter_all/SQSTM1_(8878)_Dep_gt_01.csv'
]

# Read the second file and select columns with matching names
second_file = '/content/drive/MyDrive/Colab Notebooks/DeepDEP/filtered_arranged_table.csv'
df_second = pd.read_csv(second_file)

# Iterate over the first files and generate corresponding output files
for file_path in first_file_paths:
    # Read the first file and extract the row names as a list
    df_first = pd.read_csv(file_path)
    row_names = list(df_first.iloc[:, 0])

    # Filter columns from the second file based on matching row names
    columns_to_keep = ['DepMap_ID'] + [col for col in df_second.columns if col in row_names]+['EIF2AK3 (9451) Dep','MYC (4609) Dep','NQO1 (1728) Dep','SF3B1 (23451) Dep','SQSTM1 (8878) Dep']
    df_filtered = df_second[columns_to_keep]

    # Extract the gene name from the first file path
    gene_name = file_path.split('/')[-1].split('_')[0]

    # Save the modified second file as the corresponding output file
    output_file = f'/content/drive/MyDrive/Colab Notebooks/correlation_table_filter_all/{gene_name}_gt_01.csv'
    df_filtered.to_csv(output_file, index=False)

    print(f"Saved filtered DataFrame for {gene_name} to {output_file}")
```
